```python
rank_se = secon.iloc[-3]
rank_di = secon.iloc[-4]
rank_di2 = secon.iloc[-1]
```

```python
rank_df = pd.DataFrame(index = rank_se.index)
rank_df["total"] = (rank_se + rank_di)

rank_df["total rank"] = rank_df["total"].rank()
rank_df = rank_df.sort_values(by = "total rank")

rank_df
```
